# **Visualizing World Cup Data With Seaborn**


 For this project I have explored data from the [Fifa World Cup](https://en.wikipedia.org/wiki/FIFA_World_Cup) from 1930-2014 to analyze trends and discover insights about the world’s game, fútbol!

This Fifa World Cup data is from [Kaggle](https://www.kaggle.com/abecklas/fifa-world-cup). Kaggle is a platform for data science competitions that hosts many datasets online.

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

## **Visualization of Raw Data**


```python
df = pd.read_csv('WorldCupMatches.csv')
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
df.head()
```

```python
df.info()
```

## **Bar Plot**
  #### I have created a bar chart visualizing how many goals were scored each year the World Cup was held between 1930-2014.

```python
sns.set_style('whitegrid')
sns.set_context('poster', font_scale = 0.8)
f, ax = plt.subplots(figsize=(19, 7))
sns.barplot(data = df, x = 'Year', y = 'Total Goals')
plt.title("Average Number Of Goals Scored In World Cup Matches By Year")
plt.show()
```

## **Goals Data Frame**


```python
df_goals = pd.read_csv("goals.csv")
df_goals.head()
```

## **Boxplot**
#### Now I have created a box plot to visualize the distribution of the goals data instead of just the average with a bar chart.

```python
sns.set_context('notebook', font_scale = 1.25)
f, ax3 = plt.subplots(figsize=(19, 7))
sns.set_palette('Spectral')
sns.boxplot(data = df_goals, x = 'year', y = 'goals')
plt.title('Goals Data Distribution')
plt.show()
```

## Violin Plot

#### To Visualize univariate data distribution along with Interquartile Range

```python
sns.set_context('notebook', font_scale = 1.25)
f, ax2 = plt.subplots(figsize=(19, 7))
sns.set_palette('Spectral')
sns.violinplot(data = df_goals, x = 'year', y = 'goals')
plt.title('Goals Data Distribution (KDE & Boxplot Outlining)')
plt.show()
```
