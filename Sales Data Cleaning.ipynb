## Data Cleaning

```python
daily_sales_replaced = daily_sales.replace(";,;", "-" )


daily_transactions = daily_sales_replaced.split(",")



daily_transactions_split = [c.split("-") for c in daily_transactions]

transaction_clean = []
for lst in daily_transactions_split:
  for i in lst:
    transaction_clean.append(i.strip())
```

## Data Management

```python
customers = [lst[0].strip() for lst in daily_transactions_split]

sales = [lst[1].strip() for lst in daily_transactions_split]

thread_sold = [lst[2].strip() for lst in daily_transactions_split]
```

## Total Value of Days Sales

```python
total_sales = 0
for sale in sales:
  total_sales += float(sale.replace("$", ""))

print(total_sales)
```

## Thread of any specific color was sold

```python
thread_sold_split = []
for color in thread_sold:
  if "&" in color:
    index = 0
    thread_sold_split.append(color.split("&")[index])
    index += 1
    thread_sold_split.append(color.split("&")[index])
  else :
    thread_sold_split.append(color)


def color_count(color):
  count = 0
  for c in thread_sold_split:
    if color == c:
      count +=1
    else :
      continue
  return count


print(color_count("white"))
```

### *Colors Available*

```python
colors = []

for color in thread_sold_split:
  if color in colors:
    continue
  else :
    colors.append(color)

print(colors)
```

 ### *Each Color of Thread sold Per Day*

```python
for pos in colors:
  count = 0
  count += color_count(pos)
  print("Thread Shed sold {count} threads of {pos} thread today".format(count = count, pos = pos))
```
